```python
import datetime as dt
```

```python
endDate = dt.datetime.now()
startDate = endDate - dt.timedelta(days=365)
print(startDate,endDate)
```

```python
stock= ['META', 'INTC', 'TSLA']
import yfinance as yf
df = yf.download ( stock, start = startDate, end= endDate)
```

```python
df.head()
```

```python
adjusted_close_prices= df['Adj Close']
```

```python
adjusted_close_prices.head()
```

```python
import numpy as np
```

```python
log_returns= np.log(adjusted_close_prices/adjusted_close_prices.shift(1))
print(log_returns)
cumulative_log_returns = log_returns.cumsum()
cumulative_log_returns.plot(title= 'Cumulative Returns', figsize = (10,6), )
```
